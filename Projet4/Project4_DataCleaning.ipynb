```python
# delete the columns that we are not interested in
df=df.drop(columns=['icu_id',"encounter_id","apache_3j_bodysystem", "Unnamed: 83",'apache_2_diagnosis', 'apache_3j_diagnosis', 'apache_post_operative', 'arf_apache', 'gcs_eyes_apache', 'gcs_motor_apache', 'gcs_unable_apache', 'gcs_verbal_apache','map_apache','d1_glucose_max', 'd1_glucose_min', 'd1_potassium_max', 'd1_potassium_min','aids','cirrhosis', 'diabetes_mellitus', 'hepatic_failure','immunosuppression', 'leukemia', 'lymphoma','solid_tumor_with_metastasis' ])
```

```python
df=df.drop(columns=['h1_diasbp_noninvasive_max', 'h1_diasbp_noninvasive_min', 'h1_mbp_max', 'h1_mbp_min', 'h1_mbp_noninvasive_max', 'h1_mbp_noninvasive_min','h1_sysbp_noninvasive_max', 'h1_sysbp_noninvasive_min','apache_4a_hospital_death_prob', 'apache_4a_icu_death_prob' ])                                      
           
```

```python
df.shape
```

```python
# Columns used.
# age, ethnicity,apache_2_bodysystem','hospital_death, gender, hospital_death,icu_admit_source, hospital_id, d1_sysbp_max, bmi
```

```python
col1=['gender', 'ethnicity', 'icu_admit_source', 'apache_2_bodysystem']
for i in col1:
    df.loc[df[i].isna()==True, i]='unknown'
```

```python
for i in col1:
    print(df[i].isnull().sum())
```

```python
col2=['age', 'bmi','d1_sysbp_max' ]
for i in col2:
    df.loc[df[i].isna()==True, i]= round(df[i].mean())
```

```python
for i in col2:
    print(df[i].isnull().sum())
```

```python
data_cleaned=df[['hospital_id','patient_id','age', 'gender','ethnicity','bmi','d1_sysbp_max','icu_admit_source', 'apache_2_bodysystem', 'hospital_death']]
```

```python
data_cleaned
```

```python
data_cleaned.isnull().sum()
```

```python
col3=['hospital_id', 'patient_id','age', 'd1_sysbp_max','hospital_death']
for i in col3:
    data_cleaned[i]=data_cleaned[i].astype('int')
```

```python
data_cleaned['bmi']=round(data_cleaned['bmi'])
data_cleaned['bmi']=data_cleaned['bmi'].astype(int)
```
